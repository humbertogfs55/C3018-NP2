```python
# --- Predict radiant win for random teams using trained pipeline ---

import joblib
import random
import pandas as pd

# Load trained pipeline
model_path = "../models/best_pipeline.joblib"
pipeline = joblib.load(model_path)
print("✅ Loaded model from", model_path)

# Example: define hero ID range (adjust if your dataset uses a different range)
ALL_HERO_IDS = list(range(1, 124)) 

# Generate two random 5-hero teams (no overlap)
radiant_team = random.sample(ALL_HERO_IDS, 5)
dire_team = random.sample([h for h in ALL_HERO_IDS if h not in radiant_team], 5)

print("Radiant team:", radiant_team)
print("Dire team:", dire_team)

# Build input DataFrame (must match training features)
query_df = pd.DataFrame([{
    "radiant_team_sorted": sorted(radiant_team),
    "dire_team_sorted": sorted(dire_team)
}])

def prepare_query_as_trained(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    for col in ["radiant_team_sorted", "dire_team_sorted"]:
        df[col] = df[col].map(lambda v: str(v) if isinstance(v, list) else v)
    return df

query_df_prepared = prepare_query_as_trained(query_df)

pred = pipeline.predict(query_df_prepared)[0]
proba = pipeline.predict_proba(query_df_prepared)[0][1]  # radiant win probability

print(f"Predicted radiant win: {bool(pred)} (probability = {proba:.3f})")
```
