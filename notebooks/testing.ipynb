```python
import joblib
import pandas as pd
import json
from pathlib import Path

PROJECT_ROOT = Path("/home/burns/Documents/Repos/C3018-NP2")
pipeline = joblib.load(PROJECT_ROOT / "models" / "best_pipeline.joblib")

with open(PROJECT_ROOT / "models" / "metrics.json") as f:
    metrics = json.load(f)

print("Metrics summary:", metrics.get("best_model_name"), metrics.get("best_cv_f1"))

# Load a sample row to get the correct feature structure
sample_df = pd.read_csv(PROJECT_ROOT / "data" / "processed" / "matches_clean.csv", nrows=1)
print("Available features:", list(sample_df.columns))

# Create test case with your specified hero teams
radiant_team_sorted = [12, 27, 61, 87, 90]
dire_team_sorted = [31, 43, 67, 128, 136]

# Create a test row with your hero teams and default values for other features
test_input = {
    "match_id": 9999999999,  # dummy ID
    "match_seq_num": 9999999999,  # dummy sequence
    "start_time": 1760980424,  # sample timestamp
    "duration": 1000,  # 16.67 minutes (typical game)
    "lobby_type": 7,  # ranked match
    "game_mode": 22,  # all pick
    "avg_rank_tier": 50.0,  # average rank
    "num_rank_tier": 5,  # number of ranked players
    "cluster": 182,  # sample cluster
    "radiant_team": radiant_team_sorted,  # your radiant team
    "dire_team": dire_team_sorted,  # your dire team
    "start_datetime": "2025-10-20 17:13:44+00:00",  # sample datetime
    "duration_minutes": 16.67,  # duration in minutes
    "radiant_win_int": 1,  # dummy target (will be dropped)
    "radiant_team_sorted": radiant_team_sorted,  # sorted radiant team
    "dire_team_sorted": dire_team_sorted,  # sorted dire team
    "radiant_team_size": 5,  # team size
    "dire_team_size": 5,  # team size
}

# Convert to DataFrame
df = pd.DataFrame([test_input])

# Drop the target column (radiant_win_int) as it's not used for prediction
X = df.drop(columns=["radiant_win_int"])

print("Input features:")
print(X.iloc[0])

# Make prediction
pred = pipeline.predict(X)
pred_proba = pipeline.predict_proba(X)

print(f"\nPrediction: {'Radiant Win' if pred[0] == 1 else 'Dire Win'}")
print(f"Confidence: {pred_proba[0][pred[0]]:.3f}")
print(f"Radiant Win Probability: {pred_proba[0][1]:.3f}")
print(f"Dire Win Probability: {pred_proba[0][0]:.3f}")
```
